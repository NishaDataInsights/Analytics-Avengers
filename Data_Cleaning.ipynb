# 1️⃣ **Generate histograms for numerical columns in each dataset**


```python
for name, df in dataframes.items():
    numeric_cols = df.select_dtypes(include=['number']).columns
    if len(numeric_cols) > 0:
        df[numeric_cols].hist(figsize=(12, 8), bins=30, edgecolor="black")
        plt.suptitle(f"Histograms for {name}", fontsize=14)
        plt.show()
```

🔍 **Insights from the Visualizations**
📊 **Histograms for Numerical Columns (Per Dataset)**

1️⃣ **Stock Price Distribution:**
If histograms are skewed right, it suggests some stocks have very high closing prices compared to the majority.
If they are normally distributed, it means stock prices are more balanced and stable.

2️⃣ **Volume Distribution:**
If trading volume has high peaks, some companies may have frequent spikes in investor activity.
Low trading volume may indicate less liquidity or investor interest in certain stocks.

2️⃣️**Presence of High-Priced Stocks:**

# 5️⃣ **Box Plot: Identifying Outliers in Stock Prices**

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=combined_df, x="Company", y="Close", palette="coolwarm")
plt.title("Stock Price Distribution by Company")
plt.xlabel("Company")
plt.ylabel("Closing Price")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

📦 **Box Plot: Identifying Outliers in Stock Prices**

1️⃣️**Outliers Detection:**
If some companies have extreme high/low stock prices, they may have unusual trading patterns or major market events affecting valuation.
Consistently tight distributions suggest stable stocks with less volatility.
